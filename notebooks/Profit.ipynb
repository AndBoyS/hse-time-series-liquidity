```python
import sys
from pathlib import Path
import numpy as np
import pandas as pd

%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
%config InlineBackend.figure_format='retina'
plt.rcParams["figure.dpi"] = 70

base_dir = Path().resolve().parent
data_dir = base_dir / 'data'
sys.path.append(str(base_dir))
```

```python
data = pd.read_excel(data_dir / 'Project 2_2022.xlsx')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
data['Balance'].plot(alpha=0.7);
```

```python
from tslib.profit import calculate_profit

key_rate = 0.07
day_deposit_rate = key_rate + 0.5 / 100
night_deposit_rate = key_rate - 0.9 / 100
night_loan_rate = key_rate + 1 / 100

profit = calculate_profit(data['Balance'], data['Balance'], day_deposit_rate, night_deposit_rate, night_loan_rate)
plt.plot(profit);
```
